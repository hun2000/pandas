### iloc การเลือกเเถว เเละหลัก ตามที่ต้องการ *********************การเข้าถึงเป็นตัวเลขทั้งหมด *****************************
.iloc[[index],[columns]]  =>  .iloc  จะเอา index ขึ้นก่อน(นับลงมาตามเเนวตั้ง) select
* index ขึ้นก่อน
* index , boolean
* .iloc[slice index, slice columns]  สามารถเลือกเเบบ slice ได้ด้วย => .iloc[start:limit:step index, start:limit:step columnx] ********************** slice
* .iloc[[index, index, index,...]]  จะเลือก index ที่ต้องการ เเละจะเอามาทุก columns => .iloc[[index, index, ....], [col, col, col, ...]]  เลือก index columns ตามที่ต้องการ*************
* .iloc[[index, index, ...], start:limit:step] => select index เเละ slice columns
* .iloc[start:limit:step, [columns, columns, ...]] => slice index เเละ slect columns
* *************************** ง่ายๆถ้ามี [] เเสดงว่า select เเละถ้าไม่มีเเสดงว่า slice**************************** สามารถเอามาผสมกันได้**********************

+ array boolean  =>  df.index df.columns ถึงจะได้ออกมาเป็น array boolean
+ index รับค่า Serise ไม่ได้
